```python
import pandas as pd

# for prodcuts csv

df = pd.read_csv(open('dataset.csv', 'rb')) 

del df['Unnamed: 5']
del df['Unnamed: 8']
del df['Unnamed: 11']
del df['Unnamed: 12']
del df['crawl_timestamp']
del df['Moosha Wheat Cocoa Crispies']
df['foodtype'] = df['Foodgrains, Oil & Masala']
del df['Foodgrains, Oil & Masala']

for i in df:
  print(i)
```

```python
df.head()
```

```python
df.drop(df[df.price < 0].index, inplace=True)
df.drop(df[df.discountedPrice < 0].index, inplace=True)
df.drop(df[df.title == 'nan'].index, inplace=True)
```

```python
df.isna()
for i in df:
  df.drop(df[df[i].isna()].index, inplace=True)
```

```python
df.isna()
```

```python
df.to_csv("products.csv")
```

```python
# data cleaning for customer

df = pd.read_csv(open('customer.csv', 'rb')) 
for i in df :
  print(i)
```

```python
df.head()
```

```python
df['Payment Method'] = df['Type']
del df['Type']
del df['Customer Email']
del df['Unnamed: 7']
df['First Name'] = df['Customer Fname']
df['Last Name'] = df['Customer Lname']
del df['Customer Fname'], df['Customer Lname']
df.head()
```

```python
df.isna().sum()
len(df)
```

```python
for i in df:
  df.drop(df[df[i].isna()].index, inplace=True)
len(df)
```

```python
df['pincode'] = df['Customer Zipcode'].astype(int)
df['pincode'] = df['pincode'].astype(int)
del df['Customer Zipcode']
```

```python
print(len(df))
df.drop_duplicates(subset=['Customer Id'], keep=False, inplace=True)
print(len(df))
```

```python
df.to_csv("customer.csv")
```

```python
# data cleaning for orders

df = pd.read_csv(open('orders.csv', 'rb'),low_memory=False) 
for i in df :
  print(i)
```

```python
del df['Unnamed: 15']
df.isna().sum()
df['Sales'] = df['Sales'].astype(int)
df['Customer Id'] = df['Customer Id'].astype(int)
```

```python
# df['Sales'] = df['Sales'].astype(int)  
df['Item Id'] = df['Item Id'].astype(int)
df['Order Quantity'] = df['Order Quantity'].astype(int)
# len(df)
```

```python
del df['dateAdded'],df['dateUpdated'], df['shipping date (DateOrders)']
df['shipping date'] = df['shipping date (DateOrders)']

df['order date'] = df['order date (DateOrders)']
del df['order date (DateOrders)']

df['Customer Id'] = df['Order Customer Id']
del df['Order Customer Id'],df['Order City']

df['discount percent'] = df['Order Item Discount Rate']
del df['Order Item Discount Rate']

df['Order Total'] = df['Order Item Total']
del df['Order Item Total']

df['Item Id'] = df['Order Item Id']
del df['Order Item Id']

df['Product Price'] = df['Order Item Product Price']
del df['Order Item Product Price']

df['Order Quantity'] = df['Order Item Quantity']
del df['Order Item Quantity']

df['Profit Per Order'] = df['Order Profit Per Order']
del df['Order Profit Per Order'], df['Sales'], df['Order Item Profit Ratio']
```

```python
df.head()
```

```python
df.drop(df[df['Profit Per Order'] < 0 ].index, inplace=True)
print(len(df))
```

```python
df.isna().sum()
```

```python
print(len(df))
df.drop_duplicates(subset=['Order Id'], keep=False, inplace=True)
print(len(df))
```

```python
df.to_csv('orders.csv')
```
